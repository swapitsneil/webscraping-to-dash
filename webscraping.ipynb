```python
import csv
```

```python
with open("laptops.csv", "w", encoding="utf-8", newline="") as file:
    writer = csv.writer(file)
    writer.writerow(["Name", "Price", "Rating", "Reviews"])

    for page_num in range(1, 16):
        filename = f"pages/page{page_num}.html"

        with open(filename, encoding="utf-8") as f:
            soup = BeautifulSoup(f, "html.parser")

        for laptop in soup.find_all("div", class_="tUxRFH"):
            
            name = laptop.find("div", class_="KzDlHZ")
            price = laptop.find("div", class_="Nx9bqj")
            rating = laptop.find("div", class_="XQDdHH")
            reviews = laptop.find("span", class_="Wphh3N")

            name = name.get_text(strip=True) if name else "N/A"
            price = price.get_text(strip=True) if price else "N/A"
            rating = rating.get_text(strip=True) if rating else "N/A"
            reviews = reviews.get_text(strip=True) if reviews else "N/A"

            writer.writerow([name, price, rating, reviews])
```

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("laptops.csv")
```

```python
df.head(10)
```

```python
df.info()
```

```python
df["Reviews"]
```

```python
num = df['Reviews'].str.split(r'[&\s]', expand=True)

df['no. of ratings'] = num[0].str.replace(',', '').astype(int)
df['no. of reviews'] = num[2].str.replace(',', '').astype(int)
```

```python
df['New price'] = df['Price'].str.replace('₹', '').str.replace(',', '').astype(int)
```

## Price Analysis

```python
print(f"Average Price: ₹{df['New price'].mean():.2f}")
print(f"Most Expensive: ₹{df['New price'].max()}")
print(f"Cheapest: ₹{df['New price'].min()}")
print(f"Price Range: ₹{df['New price'].max() - df['New price'].min()}")
```

## Rating Analysis

```python
print(f"Average Rating: {df['Rating'].mean():.2f}/5")
print(f"Highest Rated: {df['Rating'].max()}/5")
print(f"Lowest Rated: {df['Rating'].min()}/5")
```

## Popularity Analysis

```python
print(f"Total Reviews: {df['no. of reviews'].sum()}")
print(f"Total Ratings: {df['no. of ratings'].sum()}")
```

```python
df['Price Category'] = pd.cut(df['New price'], 
                             bins=[0, 20000, 40000, 60000, df['New price'].max()], 
                             labels=['Low Budget', 'Mid-Range', 'High Budget', 'Expensive'])

print(df['Price Category'])
```

## Popularity Score (combining ratings and review counts)

```python
df['Popularity Score'] = (df['Rating'] * df['no. of ratings']) + (df['no. of reviews'] * 0.5)
df['Popularity Score']
```

## Value for Money Score

```python
df['Value Score'] = (df['Rating'] * 1000) / (df['New price'] / 1000)
df['Value Score']
```

## Engagement Ratio (reviews vs ratings)

```python
df['Engagement Ratio'] = df['no. of reviews'] / df['no. of ratings'].replace(0, 1)
df['Engagement Ratio']
```

## Price per Rating Point

```python
df['Price per Rating'] = df['New price'] / df['Rating'].replace(0, 1)
df['Price per Rating']
```

## Relative Price Position

```python
df['Price Position'] = (df['New price'] - df['New price'].mean()) / df['New price'].std()
df['Price Position']
```

## Affordability Score (inverse relationship between price and ratings)

```python
df['Affordability Score'] = (df['Rating'] * 10) / (df['New price'] / 10000)
df['Affordability Score']
```

## Dynamic price categories using percentiles

```python
price_25 = df['New price'].quantile(0.25)
price_50 = df['New price'].quantile(0.50)
price_75 = df['New price'].quantile(0.75)

df['Price Category'] = pd.cut(df['New price'], 
                             bins=[0, price_25, price_50, price_75, df['New price'].max()], 
                             labels=['Budget', 'Mid-Range', 'Premium', 'Luxury'])

print(df['Price Category'].value_counts())
```

```python
df['Competitive_Intensity'] = abs(df['New price'] - df.groupby('Price Category', observed=True)['New price'].transform('mean'))
```
